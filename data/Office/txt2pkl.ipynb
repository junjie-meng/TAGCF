```python
import pickle
train_file_path = '/home1/workspace/mengjunjie/lightgcn/data/Office/train.txt'
test_file_path = '/home1/workspace/mengjunjie/lightgcn/data/Office/test.txt'
val_file_path = '/home1/workspace/mengjunjie/lightgcn/data/Office/val.txt'
```

```python
def transfer_txt2pkl(file_path):
    user_list = []
    item_list = []
    with open(file_path, 'r') as f:
        for data in f.readlines():
            user_item_list = data.strip().split(' ')
            user = int(user_item_list[0])
            items = [int(item) for item in user_item_list[1:]]
            for item in items:
                user_list.append(user)
                item_list.append(item)
    user_np = np.array(user_list)
    item_np = np.array(item_list)
    value_np = np.ones_like(user_np)
    print(f'num of users: {user_np.max()+1}, num of items: {item_np.max()+1}')
    mat = coo_matrix((value_np, (user_np, item_np)), shape=(user_np.max()+1, item_np.max()+1)).astype(np.float32)
    return mat
trn_mat = transfer_txt2pkl(train_file_path)
tst_mat = transfer_txt2pkl(test_file_path)
val_mat = transfer_txt2pkl(val_file_path)
```

```python
output_folder_path = '/home1/workspace/mengjunjie/RLMRec/data/Office'
with open(f'{output_folder_path}/trn_mat.pkl', 'wb') as fs:
    pickle.dump(trn_mat, fs)
with open(f'{output_folder_path}/tst_mat.pkl', 'wb') as fs:
    pickle.dump(tst_mat, fs)
with open(f'{output_folder_path}/val_mat.pkl', 'wb') as fs:
    pickle.dump(val_mat, fs)
print('done')
```
