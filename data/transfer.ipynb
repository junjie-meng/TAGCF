```python
import numpy as np
import pickle
from scipy.sparse import coo_matrix, csr_matrix
import json
from collections import Counter

attr_path = '/home1/workspace/mengjunjie/lightgcn/data/yelp/preference.txt'
# id2user_path = '/home1/workspace/mengjunjie/lightgcn/data/amazon-book/id2user.json'
# id2item_path = '/home1/workspace/mengjunjie/lightgcn/data/amazon-book/id2item.json'
output_path = 'yelp/attr_edges.pkl'
attr_edges = []
attr_counter = Counter()
with open(attr_path, 'r') as f:
    for line in f:
        item, attr = line.split(' ')
        attr_edges.append((int(item), int(attr)))
        attr_counter[int(attr)] += 1
```

```python
# filter
print(len(attr_edges))
# filtered_attr_edges = [(item, attr) for item, attr in attr_edges if attr_counter[attr] >= 15 and attr_counter[attr] <= 3000]
# attr_edges = filtered_attr_edges
# print(len(attr_edges))
```

```python

print(attr_edges[:5])  # Display first 5 edges for verification
## transfer to coo_matrix 
item_ids, attr_ids = zip(*attr_edges)

item_ids = np.array(item_ids)
attr_ids = np.array(attr_ids)
user_item_num = max(item_ids) + 1
# with open(id2item_path, 'r') as f1,\
#      open(id2user_path, 'r') as f2 :
#     id2item = json.load(f1)
#     id2user = json.load(f2)
# user_item_num = len(id2user) + len(id2item)

print(user_item_num)
print(max(attr_ids)+1)

attr_edges = coo_matrix((np.ones(len(item_ids)), (item_ids, attr_ids)), shape=(user_item_num, max(attr_ids) + 1))
with open(output_path, 'wb') as f:
    pickle.dump(attr_edges, f)
```
